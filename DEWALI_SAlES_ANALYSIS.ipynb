### EXPLORATORY DATA ANALYSIS

### Gender

```python
gender_group=df.groupby('Gender')
print(gender_group.size())


```

```python
ax = sns.countplot(x='Gender',data=df)
for bars in ax.containers:
    ax.bar_label(bars)
```

From the above graph we can see the most numbers of orders in done by Females which is 7832

```python
# Cheking which gender spent how much money
df.groupby(['Gender'], as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)


sales_gen = df.groupby(['Gender'], as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)

ax = sns.barplot(x='Gender',y='Amount',data=sales_gen)
for bars in ax.containers:
    ax.bar_label(bars,fmt="%.0f")
    
```

From the above graphs we can see that most of the buyers are females and purchasing power of Females is greater than the males

### Age

```python
plt.title("Count of Customers by Age Group")
ax = sns.countplot(x='Age Group',data=df)
```

```python
# getting it in basis of gender and age group
plt.title("Count of Customers by Age Group and Gender")
ax = sns.countplot(x='Age Group',hue="Gender",data=df)
for bars in ax.containers:
    ax.bar_label
```

From the above graph, we can see that most number of orders is done by customers of age group 26-35 when female ordered more than males

```python
# Total amount vs age group
sales_age = df.groupby(['Age Group','Gender'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)

plt.title("Total Amount Spent by Age Group and Gender")
ax=sns.barplot(x='Age Group',y='Amount',hue='Gender',data=sales_age)
for bars in ax.containers:
    ax.bar_label(bars,fmt='%.0f')
```

From the above graph, we can see that most amount of money is spent by the age group 26-35 where females spent more than double amount of males.

States

```python
# States wise no. of orders
state_order = df.groupby(["State"],as_index=False)['Orders'].sum().sort_values(by="Orders",ascending=False)

plt.figure(figsize=(15.5, 6))
plt.title("State-wise Orders")
ax= sns.barplot(x="Orders",y="State",data=state_order)
for bars in ax.containers:
    ax.bar_label(bars)
```

Above graph shows that, most number of orders are from Uttar Pradesh , Maharashtra , Karnataka and least number of orders are from Telangana , Punjab , Rajasthan.

```python
# Total Amount spend by  States
state_amount = df.groupby(["State"],as_index=False)['Amount'].sum().sort_values(by="Amount",ascending=False)

plt.figure(figsize=(30, 10))
ax = sns.barplot(x="State",y="Amount",data=state_amount)
for bars in ax.containers:
    ax.bar_label(bars,fmt="%.0f")
    
```

Above graph shows that, most amount of money generated from Uttar Pradesh which is around 1 crore 93 lakhs and least amount of money generated from Telangana which is around 11 lakhs.

### Occupation

```python
sns.set(rc={'figure.figsize':(20,5)})
ax=sns.countplot(data=df,x="Occupation")
for bars in ax.containers:
    ax.bar_label(bars,fmt='%.0f')
```

Here we can see that most numbers of order is done by IT sector,Healthcare, Aviation and least are from Agriculture, Textile,Construction.

```python
# Amount spent by each occupation
occ_amt = df.groupby(["Occupation"],as_index=False)["Amount"].sum().sort_values(by="Amount",ascending=False)
print(occ_amt)
plt.title("Amount spent per Occupation")
ax=sns.barplot(data=occ_amt,x="Occupation",y="Amount")
for bars in ax.containers:
    ax.bar_label(bars,fmt='%.0f')
```

Here we can see most amount spent by IT sector and least by Agriculture.

Products

```python
# Count of ordered Products
sns.set(rc={'figure.figsize':(27,10)})
plt.title("Products Order Count")
ax=sns.countplot(data=df,x="Product_Category")
for bars in ax.containers:
    ax.bar_label(bars)
```

Here we can see that most ordered product categories are Clothing and Apparel->Food->Electorincs->gadgets and least ordered product categories are Hand and Power tools->Tupperware->Veterinary.

```python
# Total amount spent on each product
sns.set(rc={"figure.figsize":(35,7)})
pdt_amt = df.groupby(["Product_Category"],as_index=False)["Amount"].sum().sort_values(by="Amount",ascending=False)
print(pdt_amt)


plt.title("Amount Spent in each Product")
ax = sns.barplot(data=pdt_amt,x="Product_Category",y="Amount")
for bars in ax.containers:
    ax.bar_label(bars,fmt="%.0f")
```

In above graph, we can see the most number of amount is generated from Food->Clothing->Electroincs and gadgets.

```python
# TOP 10 ORDERED PRODUCT ID
sns.set(rc={"figure.figsize":(15,5)})
pd_id = df.groupby(['Product_ID'],as_index=False)["Orders"].sum().sort_values(by="Orders",ascending=False).head(10)
print(pd_id)

ax =sns.barplot(data=pd_id,x="Product_ID",y="Orders")
for bar in ax.containers:
    ax.bar_label(bar,fmt='%.0f')

```

In above graph we can see the most ordered product id is P00265242 having 127 orders.

### Spending Behavior → Avg. Order Value by Segment

```python
# Average spend by Gender
gender_spend = df.groupby('Gender')['Amount'].mean()

# Average spend by Age Group
age_spend = df.groupby('Age Group')['Amount'].mean()

print("Average Spend by Gender:\n", gender_spend)
print("\nAverage Spend by Age Group:\n", age_spend)

# Visualization
plt.figure(figsize=(8,4))
sns.barplot(x=age_spend.index, y=age_spend.values)
plt.title("Average Spend per Customer by Age Group")
plt.ylabel("Avg. Order Value")
plt.show()
```

Here, we can see the average money spent by diff age groups

### Customer Segmentation → Age + Income + Product

```python
# Group customers by Age & Occupation
segment = df.groupby(['Age Group','Occupation'])['Amount'].sum().reset_index()

plt.figure(figsize=(12,6))
sns.barplot(x='Age Group', y='Amount', hue='Occupation', data=segment)
plt.title("Customer Segmentation by Age & Occupation")
plt.show()
```

Here we can see the age group and occupation of the customers.

### Retention Analysis → Repeat Customers

```python
# Count number of orders per Customer ID
repeat = df.groupby('User_ID')['Amount'].count().reset_index()
repeat.rename(columns={'Amount':'Orders'}, inplace=True)

# How many are repeat customers?
repeat_count = repeat['Orders'].value_counts()

plt.figure(figsize=(8,4))
sns.barplot(x=repeat_count.index, y=repeat_count.values)
plt.title("Customer Retention (Repeat Orders)")
plt.xlabel("Number of Orders by a Customer")
plt.ylabel("Count of Customers")
plt.show()
```

Above graph shows the Customers who orders repeatedly
